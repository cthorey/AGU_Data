```python
import numpy as np
from scipy.misc import imread, imresize
import matplotlib.pylab as plt

%matplotlib inline
```

```python
img = imread('kitten.jpg')
img = np.sum(img,axis=2)
```

```python
plt.imshow(img)
```

```python
U,S,V = np.linalg.svd(img,full_matrices=0)
```

```python
U.shape,S.shape,V.shape
```

```python
def lower_rank(img,k):
    U,S,V = np.linalg.svd(img,full_matrices=0)
    imgn = np.dot(U[:,:k],np.dot(np.diag(S)[:k,:k],V[:k,:]))
    plt.imshow(imgn)
```

```python

lower_rank(img,50)
```

```python
def rank(X,k):
    U,S,V = np.linalg.svd(X,full_matrices=0)
    return np.dot(U[:,:,:k],np.dot(np.diag(S)[:k,:k],V[:,:k].T))
```

```python
plt.imshow(rank(img,200))
```
